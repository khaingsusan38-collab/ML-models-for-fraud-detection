## 4. Target Variable — Fraud Distribution

```python
fraud_counts = df['fraud_bool'].value_counts()
fraud_pct    = df['fraud_bool'].value_counts(normalize=True) * 100

print(f'Non-Fraud (0): {fraud_counts[0]:>10,}  ({fraud_pct[0]:.2f}%)')
print(f'Fraud     (1): {fraud_counts[1]:>10,}  ({fraud_pct[1]:.2f}%)')
print(f'Imbalance ratio: {fraud_counts[0]/fraud_counts[1]:.0f}:1')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
colors = ['#2196F3', '#F44336']

axes[0].bar(['Non-Fraud', 'Fraud'], fraud_counts.values, color=colors, edgecolor='white', width=0.5)
axes[0].set_title('Fraud vs Non-Fraud Count', fontweight='bold')
axes[0].set_ylabel('Number of Records')
for i, v in enumerate(fraud_counts.values):
    axes[0].text(i, v + 5000, f'{v:,}', ha='center', fontsize=10)

axes[1].pie(fraud_counts.values, labels=['Non-Fraud (98.9%)', 'Fraud (1.1%)'],
            colors=colors, autopct='%1.2f%%', startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Fraud Proportion', fontweight='bold')

plt.suptitle('Class Distribution — Highly Imbalanced Dataset', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('fraud_distribution.png', bbox_inches='tight')
plt.show()
print('⚠️  Highly imbalanced — SMOTE or class_weight needed')
```

## The dataset is **highly imbalanced**: 98.9% non-fraud vs 1.1% fraud (approximately 89:1 ratio). This is realistic for real-world bank account fraud data (Jesus et al., 2022).

## 5. Numerical Features Distribution

```python
num_cols = df.select_dtypes(include='number').columns.drop('fraud_bool').tolist()

cols_per_row = 4
rows = (len(num_cols) + cols_per_row - 1) // cols_per_row
fig, axes = plt.subplots(rows, cols_per_row, figsize=(18, rows * 3.5))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=40, color='#5C6BC0', edgecolor='white', alpha=0.85)
    axes[i].set_title(col, fontsize=10, fontweight='bold')

for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Distribution of Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('numerical_distributions.png', bbox_inches='tight')
plt.show()
```

## 6. Fraud vs Non-Fraud — Key Feature Comparison

```python
key_features = ['income', 'customer_age', 'credit_risk_score',
                'velocity_6h', 'velocity_24h', 'device_fraud_count',
                'session_length_in_minutes', 'proposed_credit_limit']

fraud_df     = df[df['fraud_bool'] == 1]
non_fraud_df = df[df['fraud_bool'] == 0]

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()

for i, col in enumerate(key_features):
    axes[i].hist(non_fraud_df[col], bins=40, alpha=0.6, color='#2196F3', label='Non-Fraud', density=True)
    axes[i].hist(fraud_df[col],     bins=40, alpha=0.6, color='#F44336', label='Fraud',     density=True)
    axes[i].set_title(col, fontweight='bold', fontsize=10)
    axes[i].legend(fontsize=8)

plt.suptitle('Fraud vs Non-Fraud — Key Feature Distributions', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('fraud_feature_comparison.png', bbox_inches='tight')
plt.show()
```

## Comparing the distribution of key features between fraud and non-fraud cases reveals several notable patterns:

- **`income`**: Fraudulent applicants tend to cluster at lower income deciles, suggesting that synthetic or stolen identities may be associated with lower declared income.
- **`credit_risk_score`**: Fraud cases show a broader and lower distribution of credit risk scores, indicating that fraudulent applications are more likely to have poor credit histories — or artificially low scores due to synthetic identity data.
- **`velocity_6h`**: Fraud cases show noticeably higher values, consistent with the hypothesis that fraudsters submit multiple applications in rapid succession (burst behaviour) within short time windows.
- **`device_fraud_count`**: Fraud cases are strongly associated with higher counts, meaning that devices previously linked to fraud are significantly more likely to be used again — this is one of the strongest fraud signals in the dataset.
- **`session_length_in_minutes`**: Non-fraud applicants tend to spend more time completing the form, which may reflect genuine engagement. Fraudulent applications may be completed faster (automated or scripted behaviour).

## 7. Correlation Matrix

```python
corr_matrix = df[num_cols + ['fraud_bool']].corr()

plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, vmin=-1, vmax=1,
            linewidths=0.5, annot_kws={'size': 7})
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', bbox_inches='tight')
plt.show()
```

```python
fraud_corr = corr_matrix['fraud_bool'].drop('fraud_bool').sort_values(key=abs, ascending=False)
print('Top 10 features correlated with fraud_bool:')
print(fraud_corr.head(10).round(4))
```

## The correlation matrix reveals the linear relationships between features and the target variable (`fraud_bool`):

**Strongest positive correlations with fraud:**
- `velocity_6h` and `velocity_24h` — confirming that high application frequency in short time windows is a strong fraud indicator. This supports the inclusion of velocity features in all three models.
- `device_fraud_count` — devices with prior fraud history are strongly predictive of new fraud.

**Strongest negative correlations with fraud:**
- `credit_risk_score` — higher credit scores are associated with lower fraud probability, which is intuitive.
- `bank_months_count` — longer banking history correlates with lower fraud risk, consistent with the idea that synthetic/new identities lack established banking records.

## 8. Categorical Features — Fraud Rate

```python
cat_cols = ['payment_type', 'employment_status', 'housing_status', 'source', 'device_os']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    fraud_rate = df.groupby(col)['fraud_bool'].mean().sort_values(ascending=False) * 100
    colors_bar = plt.cm.RdYlGn_r(fraud_rate.values / fraud_rate.max())
    bars = axes[i].bar(fraud_rate.index, fraud_rate.values, color=colors_bar, edgecolor='white')
    axes[i].set_title(f'Fraud Rate by {col}', fontweight='bold')
    axes[i].set_ylabel('Fraud Rate (%)')
    for bar, val in zip(bars, fraud_rate.values):
        axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                     f'{val:.2f}%', ha='center', fontsize=9)

axes[5].set_visible(False)
plt.suptitle('Fraud Rate by Categorical Features', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('categorical_fraud_rates.png', bbox_inches='tight')
plt.show()
```

### Fraud rates vary meaningfully across categorical feature categories:

- **`housing_status`**: Certain housing categories show markedly higher fraud rates (e.g. the highest-risk category reaches ~3.75%). This may indicate that applicants in temporary or unstable housing situations present higher fraud risk — or that fraudsters are more likely to declare specific housing statuses.
- **`employment_status`**: Some employment categories show elevated fraud rates, suggesting that declared employment type may be a useful discriminating feature.
- **`device_os`**: Differences in fraud rates across operating systems may reflect the device types commonly used for automated/scripted applications.
- **`source`**: Applications submitted via INTERNET vs TELEAPP show different fraud rates, which may reflect differences in verification processes across channels.

These findings justify retaining all categorical features (after label encoding) in the model feature set.

## 9. Velocity Features — Boxplot Analysis

```python
velocity_cols = ['velocity_6h', 'velocity_24h', 'velocity_4w']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for i, col in enumerate(velocity_cols):
    data = [df[df['fraud_bool']==0][col], df[df['fraud_bool']==1][col]]
    bp = axes[i].boxplot(data, labels=['Non-Fraud', 'Fraud'], patch_artist=True,
                         medianprops=dict(color='#F44336', linewidth=2))
    bp['boxes'][0].set_facecolor('#90CAF9')
    bp['boxes'][1].set_facecolor('#EF9A9A')
    axes[i].set_title(col, fontweight='bold')
    axes[i].set_ylabel('Value')

plt.suptitle('Velocity Features — Fraud vs Non-Fraud', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('velocity_boxplots.png', bbox_inches='tight')
plt.show()
```

### The boxplots compare the distribution of velocity features between fraud and non-fraud cases:

- **`velocity_6h` (Non-Fraud mean ≈ 5,000)**: The relatively high mean for non-fraud cases reflects normal application activity across all users sharing similar devices or IP ranges. However, fraud cases show a **wider spread and higher outliers**, consistent with burst-application behaviour.

- **`velocity_24h` (Non-Fraud mean ≈ 4,500)**: Notably lower than both the 6h and 4w values for non-fraud cases. A possible explanation is the **daily activity cycle** — the 24h window captures both peak and off-peak hours, diluting the per-hour rate compared to the 6h window (which may capture only peak hours). This warrants further investigation.

- **`velocity_4w` (Non-Fraud mean slightly below 5,000)**: The 4-week window smooths out short-term burst behaviour. For fraud cases, the 4w value is relatively lower than the 6h value — this supports the hypothesis that **fraudulent activity tends to be concentrated in short bursts** rather than sustained over weeks, making short-window velocity (especially `velocity_6h`) a stronger discriminating feature.
